```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
#Open firefox and go to url
driver = webdriver.Firefox()
driver.get('https://www.bolsamadrid.es')

driver.title
```

```python
#Get a list of all menus with links
market = driver.find_elements(By.CLASS_NAME,
                             'MenuLink')
```

```python
#Returns the index of a clickable submenu. If the submenu is not found, raises an exception 
def getSubmenuIndex(market, menuLabel):
    
    submenuIndex = -1
        
    for count, element in enumerate(market):
        if element.text == menuLabel:
            submenuIndex = count
        
    if submenuIndex < 0:
        raise Exception(f'Submenu "{menuLabel}" is not found')
            
    return submenuIndex
```

```python
#Click on submenu Acciones. Case sensitive
submenuLabel = 'Acciones'
market[getSubmenuIndex(market, submenuLabel)].click()
driver.implicitly_wait(10)
```

```python
#Get the table of the actions 
table = driver.find_element(By.ID, 'ctl00_Contenido_tblAcciones')
```

```python
def makeRows(tableRows):
    rows = []
    #Get each element and convert it to text
    for row in tableRows[1:]:
    
        rowList = row.find_elements(By.XPATH, './/td')

        rowElements = [element.text for element in rowList]

        rows.append(rowElements)
        
    return rows
```

```python
#Get each row of the table
tableRows = table.find_elements(By.XPATH, './/tbody//tr')

rows = makeRows(tableRows)
rows[5]
```

```python
#Get the link to each individual enterprise
linkElements = table.find_elements(By.XPATH, './/tbody//tr//td//a')

links = [element.get_attribute('href') for element in linkElements]
```

```python
def appendLinks(rows, links):
    
    ultimateRows = []
    
    for row, link in zip(rows, links):
        row.append(link)
        
    return rows
```

```python
ultimateRows = appendLinks(rows[1:], links)
ultimateRows[7]
```

```python
def toNumbers(rows):
    
    for index, row in enumerate(rows):
        for count, element in enumerate(row[1:7]):
            elementNoMilPoint = element.replace('.', '')
            elementDecimalPoint = elementNoMilPoint.replace(',' , '.' )
            row[count + 1] = float(elementDecimalPoint)
        
        rows[index] = row
    return rows
```

```python
ultimateRows = toNumbers(ultimateRows)
ultimateRows[8]
```

```python
headers = ['Nombre', 'Últ.','% Dif.', 'Máx.', 'Mín.', 
           'Volumen', 'Efectivo (miles €)', 'Fecha', 'Hora', 'Link']

scrappedDf = pd.DataFrame(ultimateRows, columns = headers)
```
